# **Exploring the Dataset with Directed Questions**

```python
#Q1: You'd like to get the first and last name of the last member(s) who signed up. Try not to use the LIMIT clause for your solution.
query_last_member = """SELECT m.firstname, m.surname, m.joindate
FROM members AS m
WHERE m.joindate = (SELECT MAX(m.joindate) FROM members as m)
"""

df_last_member = pd.read_sql(query_last_member, conn)
print(df_last_member)
```

```python
#Q2: Produce a list of all members who have used a tennis court.
#Include in your output the name of the court, and the name of the member formatted as a single column. Ensure no duplicate data, and order by the member name.
query_tennis_courts = """SELECT m.firstname, m.surname
                    FROM members as m 
                    WHERE m.memid IN (SELECT DISTINCT b.memid FROM bookings as b WHERE b.facid = 0)"""

df_tennismembers = pd.read_sql(query_tennis_courts, conn)
print(df_tennismembers)
```

```python
#Q3: Produce a report of members and who recommended them in alphabetic surname,firstname order
query_recommended_by = """SELECT m.firstname, m.surname, r.firstname AS recommended_by_firstname, r.surname AS recommended_by_surname
                    FROM members AS m
                    LEFT JOIN members AS r ON m.recommendedby = r.memid
                    ORDER BY m.surname, m.firstname"""
#this is the right way because it includes all members, even those who were not recommended by anyone

df_recommended_by = pd.read_sql(query_recommended_by, conn)
print(df_recommended_by)
```

```python
#Q4: Some of the facilities charge a fee to members, but some do not. Please list the names of the facilities that do.
query_member_fee = """SELECT f.name
                    FROM facilities AS f
                    WHERE f.membercost > 0"""
df_member_fee = pd.read_sql(query_member_fee, conn)
print(df_member_fee)
```

```python
#Q6: How many facilities do not charge a fee to members?

query_no_member_fee = """SELECT f.name 
                    FROM facilities AS f
                    WHERE f.membercost = 0"""
df_no_member_fee = pd.read_sql(query_no_member_fee, conn)
print(df_no_member_fee)
```

```python
#Q7: Write an SQL query to show a list of facilities that charge a fee to members,
#where the fee is less than 20% of the facility's monthly maintenance cost.
#Return the facid, facility name, member cost, and monthly maintenance of the
#facilities in question.

query_member_fee_less = """SELECT facid, name, membercost, monthlymaintenance
                        FROM facilities
                        WHERE membercost < 0.2 * monthlymaintenance
                        AND membercost > 0""" 
df_q8 = pd.read_sql(query_member_fee_less, conn)
print(df_q8)

```

```python
#Q8: Write an SQL query to retrieve the details of facilities with ID 1 and 5.
query_facid_1_5 = """SELECT *
                    FROM facilities
                    WHERE facid IN (1, 5)"""

df_q8 = pd.read_sql(query_facid_1_5, conn)
print(df_q8)
```

```python
#Q9: Produce a list of facilities, with each labelled as 'cheap' or 'expensive', depending on if their monthly maintenance cost is more than $100.
#Return the name and monthly maintenance of the facilities in question.
query_cheap_expensive = """SELECT name, monthlymaintenance,
                        CASE WHEN monthlymaintenance > 100 THEN 'expensive' ELSE 'cheap' END AS label
                        FROM facilities
                        
                        """
df_q9 = pd.read_sql(query_cheap_expensive, conn)
print(df_q9)
```

```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue.
#Remember that there's a different cost for guests and members!
query_total_revenue = """SELECT name, 
                        CASE
                        WHEN b.memid = 0 THEN f.guestcost * b.slots
                        ELSE f.membercost * b.slots
                        END AS total_revenue
                        FROM bookings as b
                        JOIN facilities AS f ON b.facid = f.facid
                        GROUP BY f.name
                        HAVING total_revenue < 1000
                        ORDER BY total_revenue;"""

df_q10 = pd.read_sql(query_total_revenue, conn)
print(df_q10)
```

```python
#Q11: Find the facilities with their usage by member, but not guests
query_usage_by = """SELECT f.name AS facility_name, m.firstname, m.surname, SUM(b.slots) AS usage
                    FROM bookings AS b
                    JOIN facilities AS f ON b.facid = f.facid
                    JOIN members AS m ON b.memid = m.memid
                    WHERE b.memid != 0
                    GROUP BY f.name, m.firstname, m.surname
                    ORDER BY m.firstname, m.surname;"""

df_usage_by = pd.read_sql(query_usage_by, conn)
print("Facilities with their usage by member, but not guests:")
print(df_usage_by)
```

# **1. Revenue & Financial KPIs**


**Total Revenue per Facility:**

Summing booking slots multiplied by facility costs (guest vs. member rates) to determine which facilities generate the most income.

```python
query_total_revenue = """SELECT f.name AS facility_name, SUM
                        CASE
                        WHEN b.memid = 0 THEN f.guescost
                        ELSE f.membercost
                        END) AS total_revenue
                        FROM bookings AS b
                        JOIN facilities AS f ON b.facid = f.facid
                        GROUP BY f.name;
                        ORDER BY total_revenue DESC;"""

df_total_revenue = pd.read_sql(query_total_revenue, conn)

print("Total Revenue by Facility:")
print(df_total_revenue)
```

**Average Revenue per Member (ARPM):**

Calculate the total spending (fees + booking costs) per member to identify high-value users.

**Guest Fee Revenue:**
Total income generated specifically from non-member guests, indicating the popularity of the club with outside visitors.

**Facility Cost Recovery:**
Comparing the revenue generated by a facility to the frequency of its use to determine if rates need adjustment.

# **2. Usage & Operational KPIs**

**Facility Utilization Rate (Booking Density):** The ratio of booked hours to total available operating hours for facilities

**Member-to-Guest Ratio:** Tracking how many guests are brought by members to measure member satisfaction (acting as advocates).
