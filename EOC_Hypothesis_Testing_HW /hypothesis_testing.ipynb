```python
import statsmodels
```

Null Hypothesis (H₀): 
𝜇 = 30
μ=30
(The community center’s average member age is 30 years.)

Alternative Hypothesis (Hₐ): 
𝜇 ≠ 30 μ =30
(The true average member age is different from 30 years.)

Significance level: α = 0.05 (two-tailed test)

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# 1. Sample Data (Now using a larger sample for Z-test validity)
# Note: Z-tests are more reliable with n > 30
ages = np.array([28, 32, 35, 29, 31, 34, 30, 27, 33, 31, 
                 29, 30, 32, 28, 30, 31, 34, 29, 30, 32,
                 33, 27, 28, 31, 30, 32, 29, 30, 34, 31])

# 2. Set the Population Mean to test against (Null Hypothesis)
pop_mean_null = 30

# 3. Perform One-Sample Z-Test
# value: the mean under the null hypothesis
z_stat, p_val = ztest(ages, value=pop_mean_null)

# 4. Print Results
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

# 5. Decision Logic
alpha = 0.05
if p_val < alpha:
    print(f"Decision: Reject H0 (p < {alpha})")
else:
   print(f"Decision: Fail to Reject H0 (p >= {alpha})") 
```

```python
# Step 1: Add the outlier value 110
ages_with_outlier = np.append(ages, 110)

# Step 2: Run z-test again
z_stat2, p_val2 = ztest(ages_with_outlier, value=pop_mean_null)

# Step 3: Print new results
print("Z-Statistic with outlier:", round(z_stat2, 4))
print("P-Value with outlier:", round(p_val2, 4))

# Step 4: Decision
if p_val2 < alpha:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```

Task 3:When we add 110, the sample mean increased. However the standard Deviation also increased a significant amount because 110 is far away from the other values. 

Since the Z-test uses the standard error, the outlier changes both the Z value and the p value.

This shows that the Z tests are sensitive to outliers because extreme values can strongly affect the mean and variability of the data.

Task 4: 
The formula for a Z-test is: Z = (x̄ − μ) / (σ / √n) 
So, If the sample size (n) is quadrupled, then √n doubles. 

Due to the denominator (σ / √n) becoming smaller, the Z-statistic becomes larger in magnitude. When the Z-statistic becomes larger, the p-value becomes smaller. This means a larger sample size makes it easier to reject the null hypothesis because there is less uncertainty in the estimate.
